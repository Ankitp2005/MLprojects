sentiment analyses + wordcloud

```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
import matplotlib.pyplot as plt
import re
from wordcloud import WordCloud
from sklearn.naive_bayes import MultinomialNB
```

```python
data = {
    "review":[
        "product quality is excilent",
        "delivery was very late",
        "worst experience ever",
        "very happy with product",
        "poor delivery service"
    ],
    "sentiment" : [1,0,0,1,0]
}
```

```python
df = pd.DataFrame(data)
```

```python
def clean_text(text):
  text = text.lower()
  text = re.sub(r'[^a-z\s]','',text)
  return text
```

```python
df['review']= df['review'].apply(clean_text)
df.head()
```

```python
df['clean_review'] = df['review'].apply(clean_text)
```

# step 2: clean text

# step 3: convert text to numbers

```python
vectorizer = CountVectorizer()
```

```python
x = vectorizer.fit_transform(df["clean_review"])

y = df["sentiment"]

model = MultinomialNB()
model.fit(x, y)
```

# step 5: predict sentiment for new reviews

```python
# step 5 : predict sentiment for new review

new_review = ['product is good']
new_review_clean = [clean_text(new_review[0])]
new_x = vectorizer.transform(new_review_clean)
```

```python
prediction = model.predict(new_x)
print(prediction)
```

step 6 : create word cloud

```python
allText = "".join(df['clean_review'])
```

```python
wordcloud = WordCloud(
    width = 800,
    height = 400,
    background_color = 'white',
).generate(allText)
```

display word cloud

```python
plt.figure(figsize=(10,5))
plt.imshow(wordcloud , interpolation = "bilinear")
plt.axis('off')
plt.show()
```

# **nlp TEACHING APP using stramlit**

```python
import streamlit as st
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
```

```python
nltk.download('punkt')
nltk.download('stopwords')
```

```python
st.title(" simple NLP teaching app")
```

```python
text = st.text_area('enter a sentenece ',"the delivery was good but late")
if st.button('analyze Text'):
  tokens = word_tokenize(text.lower())

  stop_words = set(stopwords.words('english'))

  filtered_tokens = [w for w in tokens if w.isalpha() and w not in stop_words]

  positive_words = ['good', 'happy', 'excellent'] # Example positive words
  negative_words = ['bad', 'late', 'slow', 'worst', 'poor'] # Example negative words

  pos = sum(1 for w in filtered_tokens if w in positive_words)

  neg = sum(1 for w in filtered_tokens if w in negative_words)


  if pos> neg:
    sentiment = 'positive'
  elif neg>pos:
    sentiment = 'negative'
  else:
    sentiment = 'neutral'

  st.subheader('results')
  st.write('tokens :',tokens)
  st.write('filtered tokens: ',filtered_tokens)
  st.write('sentiment:',sentiment)
```
